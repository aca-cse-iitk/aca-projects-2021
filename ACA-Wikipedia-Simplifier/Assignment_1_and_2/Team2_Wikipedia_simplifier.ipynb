3. You start with 6 guesses

      loose two guesses for each wrong vowel guessed 

   Game ends when you run out of guesses

4. [Only For Hint Mode] When u insert `*` u get hints to what the words might be

5. When u insert a letter it  accordingly

6. Any invalid input will give u a warning   
   after three warning you start loosing guesses  
   Invalid input: any repeated letter or anything which is not letter. (`*` is allowed in hint mode)

7. Score is calculated as : 
    > `Total score = guesses_remaining * number unique letters in secret_word`

````python
# FIRST RUN PREV CODE AND GET "words.txt" :D

# Author : Salazar.h
# File   : hangman.py
# Path   : ~/Documents/Wikipedia Simplifier/ps2

# Hangman Game

import random
import string

WORDLIST_FILENAME = "./ps2/words.txt"


def load_words():
    # Returns a list of valid words. Words are strings of lowercase letters.

    print("Loading word list from file...")

    # inFile: file
    inFile = open(WORDLIST_FILENAME, 'r')

    # line: string
    line = inFile.readline()

    # wordlist: list of strings
    wordlist = line.split()

    print("  ", len(wordlist), "words loaded.")
    return wordlist


def choose_word(wordlist):

    # wordlist (list): list of words (strings)
    # Returns a word from wordlist at random

    return random.choice(wordlist)


# Load the list of words into the variable wordlist
# so that it can be accessed from anywhere in the program
wordlist = load_words()


def is_word_guessed(secret_word, letters_guessed):
    '''
    secret_word: string, the word the user is guessing; assumes all letters are lowercase

    letters_guessed: list (of letters), which letters have been guessed so far; assumes that all letters are lowercase

    returns: boolean, True if all the letters of secret_word are in letters_guessed; False otherwise
    '''
    for letter in secret_word:
        if letter not in letters_guessed:
            return False
    return True


def get_guessed_word(secret_word, letters_guessed):
    '''
    secret_word: string, the word the user is guessing

    letters_guessed: list (of letters), which letters have been guessed so far

    returns: string, comprised of letters, underscores (_), and spaces that represents 
    which letters in secret_word have been guessed so far.
    '''
    out = ""
    for letter in secret_word:
        if letter in letters_guessed:
            out += letter
        else:
            out += "_ "

    return out


def get_available_letters(letters_guessed):
    '''
    letters_guessed: list (of letters), which letters have been guessed so far
    returns: string (of letters), comprised of letters that represents which letters have not
             yet been guessed.
    '''
    out = ""
    a = string.ascii_lowercase
    for letter in a:
        if letter not in letters_guessed:
            out += letter

    return out


vowels = "aeiou"


def hangman(secret_word):

    # Hangman Driver Code
    '''
    secret_word: string, the secret word to guess.

    Starts up an interactive game of Hangman.

    * At the start of the game, let the user know how many 
      letters the secret_word contains and how many guesses s/he starts with.

    * The user should start with 6 guesses

    * Before each round, you should display to the user how many guesses
      s/he has left and the letters that the user has not yet guessed.

    * Ask the user to supply one guess per round. Make sure to check that the user guesses a letter

    * The user should receive feedback immediately after each guess 
      about whether their guess appears in the computer's word.

    * After each guess, you should display to the user the 
      partially guessed word so far.

    * If the guess is the symbol *, print out all words in wordlist that
      matches the current guessed word. 

    '''
    print("Welcome to the game Hangman without Hints!")
    print("I am thinking of a word that is {} letters long.".format(len(secret_word)))

    # warning left thoda mehnat kar lete hai

    warning_left = 2
    letters_guessed = []
    guessed_word = get_guessed_word(secret_word, letters_guessed)
    i = -1

    # guesssing loop

    while i < 5 and not is_word_guessed(secret_word, letters_guessed):
        i += 1
        print("------------------")

        # print number of guesses left
        print("You have {} guess{} left.".format(6-i, 'es' if i < 5 else ''))

        # letters available
        available_letters = get_available_letters(letters_guessed)
        print("Available letters: " + available_letters)

        # ask for a input
        letter = input("Please guess a letter : ")
        letter = letter.lower()

        # if the letter is already been guessed
        if letter in letters_guessed:
            if warning_left == -1:
                print(
                    "Oops! You've already guessed that letter. You loose a guess as no warning left: " + guessed_word)
            else:
                print("Oops! You've already guessed that letter. You now have {} warning{}: ".format(
                    warning_left, '' if warning_left == 1 else 's') + guessed_word)
                warning_left -= 1
                i -= 1
            continue

        # if not the letter
        if letter not in string.ascii_lowercase:
            if warning_left == -1:
                print(
                    "Oops! That is not a valid letter. You loose a guess as no warning left: " + guessed_word)
            else:
                print("Oops! That is not a valid letter. You now have {} warning{}: ".format(
                    warning_left, '' if warning_left == 1 else 's') + guessed_word)
                warning_left -= 1
                i -= 1
            continue

        letters_guessed += letter
        guessed_word = get_guessed_word(secret_word, letters_guessed)

        # if letter avialable or not available in it
        if letter in secret_word:
            print("Good guess: " + guessed_word)
            i -= 1
        else:
            print("Oops! That letter is not in my word: " + guessed_word)
            if letter in vowels:
                i += 1

    # the final result so just end it already
    if is_word_guessed(secret_word, letters_guessed):
        print("Congratulations, you won!")
        res = 0
        for x in letters_guessed:
            if x in secret_word:
                res += 1
        i = 5-i if i < 6 else 0
        res = res * i
        print("Your total score for this game is : {}".format(res))
    else:
        print("Sorry, you ran out of guesses. The word was {}.".format(secret_word))

    return


def match_with_gaps(my_word, other_word):
    '''
    my_word: string with _ characters, current guess of secret word
    other_word: string, regular English word
    returns: boolean, True if all the actual letters of my_word match the 
        corresponding letters of other_word, or the letter is the special symbol
        _ , and my_word and other_word are of the same length;
        False otherwise: 
    '''
    my_word = my_word.replace(" ", "")
    
    if len(my_word) != len(other_word):
        return False
    
    for i in range(len(my_word)):
        if my_word[i] == '_':
            continue
        if my_word[i] != other_word[i]:
            return False
    
    return True


def show_possible_matches(my_word):
    '''
    my_word: string with _ characters, current guess of secret word
    returns: nothing, but should print out every word in wordlist that matches my_word
             Keep in mind that in hangman when a letter is guessed, all the positions
             at which that letter occurs in the secret word are revealed.
             Therefore, the hidden letter(_ ) cannot be one of the letters in the word
             that has already been revealed.
    '''
    reslt = []
    for x in wordlist:
        if match_with_gaps(my_word, x):
            reslt.append(x)
    
    
    print("Possible word matches are: ")
    print(* reslt, sep=', ')
    print("Guess for: ", my_word)


def hangman_with_hints(secret_word):

    # Hangman with hints Driver Code

    '''
    secret_word: string, the secret word to guess.

    Starts up an interactive game of Hangman.

    * At the start of the game, let the user know how many 
      letters the secret_word contains and how many guesses s/he starts with.

    * The user should start with 6 guesses

    * Before each round, you should display to the user how many guesses
      s/he has left and the letters that the user has not yet guessed.

    * Ask the user to supply one guess per round. Make sure to check that the user guesses a letter

    * The user should receive feedback immediately after each guess 
      about whether their guess appears in the computer's word.

    * After each guess, you should display to the user the 
      partially guessed word so far.

    * If the guess is the symbol *, print out all words in wordlist that
      matches the current guessed word. 

    '''
    print("Welcome to the game Hangman with Hints!")
    print("I am thinking of a word that is {} letters long.".format(len(secret_word)))
    
    # warning left thoda mehnat kar lete hai
    
    warning_left = 2
    letters_guessed = []
    guessed_word = get_guessed_word(secret_word, letters_guessed)
    i = -1

    # guesssing loop

    while i < 5 and not is_word_guessed(secret_word, letters_guessed):
        i += 1
        print("------------------")

        # print number of guesses left
        print("You have {} guess{} left.".format(6-i, 'es' if i < 5 else ''))

        # letters available
        available_letters = get_available_letters(letters_guessed)
        print("Available letters: " + available_letters)

        # ask for a input
        letter = input("Please guess a letter : ")
        letter = letter.lower()

        # reply to hint
        if letter == '*':
            show_possible_matches(guessed_word)
            i -= 1
            continue
        
        # if the letter is already been guessed
        if letter in letters_guessed:
            if warning_left == -1:
                print(
                    "Oops! You've already guessed that letter. You loose a guess as no warning left: " + guessed_word)
            else:
                print("Oops! You've already guessed that letter. You now have {} warning{}: ".format(
                    warning_left, '' if warning_left == 1 else 's') + guessed_word)
                warning_left -= 1
                i -= 1
            continue

        # if not the letter
        if letter not in string.ascii_lowercase:
            if warning_left == -1:
                print(
                    "Oops! That is not a valid letter. You loose a guess as no warning left: " + guessed_word)
            else:
                print("Oops! That is not a valid letter. You now have {} warning{}: ".format(
                    warning_left, '' if warning_left == 1 else 's') + guessed_word)
                warning_left -= 1
                i -= 1
            continue

        letters_guessed += letter
        guessed_word = get_guessed_word(secret_word, letters_guessed)

        # if letter avialable or not available in it
        if letter in secret_word:
            print("Good guess: " + guessed_word)
            i -= 1
        else:
            print("Oops! That letter is not in my word: " + guessed_word)
            if letter in vowels:
                i += 1

    # the final result so just end it already
    if is_word_guessed(secret_word, letters_guessed):
        print("Congratulations, you won!")
        res = 0
        for x in letters_guessed:
            if x in secret_word:
                res += 1
        i = 5-i if i < 6 else 0
        res = res * i
        print("Your total score for this game is : {}".format(res))
    else:
        print("Sorry, you ran out of guesses. The word was {}.".format(secret_word))

    return


if __name__ == "__main__":

    secret_word = choose_word(wordlist)
    
    # print(secret_word) ## turn it on if u wanna cheat XD

    # ask if u want hint
    hint = "s"
    while hint  not in ['y','n']:
        hint = input("Would u like to have hints? [y/n]: ")
        hint = hint.lower()
    
    # see what function to call
    if hint == 'n':
        hangman(secret_word)
    else:
        hangman_with_hints(secret_word)

# Yaaayyy KHATAM !
````
